```python
# Note: union |  intersection &  difference -  symmetric_difference ^
# issubset <= issuperset >=    disjoint

warehouse_items={"laptop","ups","router","printers"}
store_items={"laptop","mouse","keyboard"}

missing_in_store=warehouse_items.difference(store_items)
print(missing_in_store)
missing_in_store=warehouse_items - store_items
print(missing_in_store)
```

```python
college_A={"maths","science","history","geography","maths"}
college_B={"maths","arts","science"}

all_courses=college_A.union(college_B)
print(all_courses)
all_courses=college_A | college_B
print(all_courses)

unique_courses=college_A.symmetric_difference(college_B)
print(unique_courses)

unique_courses=college_A ^ college_B
print(unique_courses)
```

```python
# use for to unpack values from iterables
list1=[10,20,30,40,50]
for num in list1:
    print(num,end=" ")
print(" ")
tuple1=(10,20,30,40,50)
for num in tuple1:
    print(num,end=" ")
print(" ")
set1={10,20,30,40,50}
for num in set1:
    print(num,end=" ")
print(" ")
states={"maharashtra":"mumbai","karnataka":"bengaluru","tamilnadu":"chennai"}
for state in states: # Note: implicit i.e. list of keys() are returned
    print(state,end=" ")
print(" ")
for state in states:
    print(states[state],end=" ")
print(" ")
for state in states.keys(): # use keys() method explicitly!
    print(state,end=" ")
print(" ")
for capitals in states.values(): # [v1,v2...]
    print(capitals,end=" ")
print(" ")
for each_record in states.items(): # [(k1,v1),(k2,v2)...] 
    print(each_record[0],each_record[1],end=" ") # unpack tuple using []
print(" ")
for state,capital in states.items(): # [(k1,v1),(k2,v2)...] # use multiple variable assingment to unpack tuple!
    print(f"{capital} is in {state}")
```

```python
# create * in function definition to create a tuple 
# which can hold variable number of arguments
def add(*arg1): # (1,2,3)
    #print(type(arg1),arg1)
    total=0
    for num in arg1:
        total+=num
    return total
    
print(add(1,2,3))
print(add())
print(add(1,2,3,4,5))
```

```python
# Note: ** will create dictionary which can hold kwarguments
def add(**kwarg1):
    print(type(kwarg1),kwarg1)

print(add(x=10,y=20)) # kwarguments
```

```python
def add(*arg1,**kwargs):
    total=0
    for num in arg1:
        total+=num
    for key,value in kwargs.items():
        total+=value
    return total
    
print(add(1,2,3))
print(add(x=100,y=200))
print(add(10,20,x=100,y=200))
print(add())
```
