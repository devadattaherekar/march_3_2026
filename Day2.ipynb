```python
# Note: union |  intersection &  difference -  symmetric_difference ^
# issubset <= issuperset >=    disjoint

warehouse_items={"laptop","ups","router","printers"}
store_items={"laptop","mouse","keyboard"}

missing_in_store=warehouse_items.difference(store_items)
print(missing_in_store)
missing_in_store=warehouse_items - store_items
print(missing_in_store)
```

```python
website_1={"leon","liam","russel","leon","leon"}
website_2={"adam","russel","leon","william"}

common_users=website_1.intersection(website_2)
print(common_users)
common_users=website_1 & website_2
print(common_users)
```

```python
college_A={"maths","science","history","geography","maths"}
college_B={"maths","arts","science"}

all_courses=college_A.union(college_B)
print(all_courses)
all_courses=college_A | college_B
print(all_courses)

unique_courses=college_A.symmetric_difference(college_B)
print(unique_courses)

unique_courses=college_A ^ college_B
print(unique_courses)
```

```python
website_1={"leon","liam","russel","leon","leon"}
website_2={"adam","russel","leon","william"}

print(website_1.issubset(website_2))
print(website_1 <= website_2)
print(website_1.issuperset(website_2))
print(website_1 >= website_2)
print(website_1.isdisjoint(website_2))
```

```python
# use for to unpack values from iterables
list1=[10,20,30,40,50]
for num in list1:
    print(num,end=" ")
print(" ")
tuple1=(10,20,30,40,50)
for num in tuple1:
    print(num,end=" ")
print(" ")
set1={10,20,30,40,50}
for num in set1:
    print(num,end=" ")
print(" ")
states={"maharashtra":"mumbai","karnataka":"bengaluru","tamilnadu":"chennai"}
for state in states: # Note: implicit i.e. list of keys() are returned
    print(state,end=" ")
print(" ")
for state in states:
    print(states[state],end=" ")
print(" ")
for state in states.keys(): # use keys() method explicitly!
    print(state,end=" ")
print(" ")
for capitals in states.values(): # [v1,v2...]
    print(capitals,end=" ")
print(" ")
for each_record in states.items(): # [(k1,v1),(k2,v2)...] 
    print(each_record[0],each_record[1],end=" ") # unpack tuple using []
print(" ")
for state,capital in states.items(): # [(k1,v1),(k2,v2)...] # use multiple variable assingment to unpack tuple!
    print(f"{capital} is in {state}")
```

```python
# Generators act as an alternative for iterables or collections
# unlike collections they do not store values ( so memory footprint is less)
# Almost 90% of python API return generator functions
# Generators will always remember its state and will resume the call after suspension
# compute will only happen when there is a call
# There are 3 ways to access values from generators
# 1. use conversion functions for testing purpose i.e. to check values from generator functions
# 2. To access all values from generators use for loops
# for loops work on both iterables as well as generators!
# 3. use next() iterator to access one value at a time
```
