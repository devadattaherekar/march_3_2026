```python
# value is reference (special variable which can only hold address of the object)
# 100 is immutable object (constant object) state cannot be modified of class int
# any operation  (arithmetic operation or functionality) on immutable object creates a new object
# at new memory location
# id() will print address of int object / it will not act on reference it acts on referent
# type() will print data type of the object
# Keywords are objects / Everything is an object in python (even functions)
# There is no compile time / everything is runtime by default!
# runtime consumes a lot of memory and will make your application slow at the cost of this 
# flexibility / maintain / reuse / easy / simple / good looks 
value=100
print(value,id(value),type(value))
newvalue=value # shallow copy is copy the same address in both the references!
value=200
print(value,id(value),type(value))
print(newvalue,id(newvalue),type(newvalue))
value+=1
print(value,id(value),type(value))
value=3.142 # There is no double in python # you only have float
print(value,id(value),type(value))
value=False
print(value,id(value),type(value))
value="False"
print(value,id(value),type(value))
value='f'
print(value,id(value),type(value))
value="""false"""
print(value,id(value),type(value))
value=(4+2j) # class complex
print(value,id(value),type(value))
value=None
print(value,id(value),type(value))
print(type(id),type(type))
```
